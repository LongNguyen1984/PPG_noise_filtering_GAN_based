# From Data to DataLoader

```python
class DataLoaders(GetAttr):
    def __init__(self, *loaders): self.loaders = loaders
    def __getitem__(self, i): return self.loaders[i]
    train,valid = add_props(lambda i,self: self[i])
```

```python
ppgs = DataBlock(
    blocks=(ImageBlock, CategoryBlock), 
    get_items=get_image_files, 
    splitter=RandomSplitter(valid_pct=0.2, seed=42),
    get_y=parent_label,
    item_tfms=Resize(128))
```

```python
dls = ppgs.dataloaders(path)
```

### Validate the ppg data

```python
dls.valid.show_batch(max_n=4, nrows=1)
```

### Resize the images

```python
ppgs = ppgs.new(item_tfms=Resize(128, ResizeMethod.Squish))
dls = ppgs.dataloaders(path)
dls.valid.show_batch(max_n=4, nrows=1)
```

```python
ppgs = ppgs.new(item_tfms=Resize(128, ResizeMethod.Pad, pad_mode='zeros'))
dls = ppgs.dataloaders(path)
dls.valid.show_batch(max_n=4, nrows=1)
```

```python
ppgs = ppgs.new(item_tfms=RandomResizedCrop(128, min_scale=0.3))
dls = ppgs.dataloaders(path)
dls.valid.show_batch(max_n=4, nrows=1)
```

### Data Augumentation

```python
ppgs = ppgs.new(item_tfms=Resize(128), batch_tfms=aug_transforms(mult=2))
dls = ppgs.dataloaders(path)
dls.train.show_batch(max_n=8, nrows=2, unique=True)
```

### Trainning the model

```python
ppgs = ppgs.new(
    item_tfms=RandomResizedCrop(256, min_scale=0.5),
    batch_tfms=aug_transforms())
dls = ppgs.dataloaders(path)
```

```python
learn = vision_learner(dls, resnet18, metrics=error_rate)
learn.fine_tune(4)
```

```python
interp = ClassificationInterpretation.from_learner(learn)
interp.plot_confusion_matrix()
```

```python
interp.plot_top_losses(5, nrows=1)
```

```python
#hide_output
cleaner = ImageClassifierCleaner(learn)
cleaner
```

```python
#hide
# for idx in cleaner.delete(): cleaner.fns[idx].unlink()
for idx,cat in cleaner.change(): 
    if Path(cleaner.fns[idx]).is_file():
        file_name = Path(cleaner.fns[idx]).stem + '_0' + '.jpg'
    shutil.move(str(cleaner.fns[idx]), path/cat/file_name)
```
